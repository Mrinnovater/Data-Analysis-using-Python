```python

age_vector = insurance_data['age'].values
bmi_vector = insurance_data['bmi'].values

sex_factor = pd.Categorical(insurance_data['sex'])
smoker_factor = pd.Categorical(insurance_data['smoker'])

example_list = [age_vector[:5], bmi_vector[:5], list(sex_factor[:5])]

matrix_example = np.array([age_vector[:5], bmi_vector[:5]])

subset_df = insurance_data[['age', 'bmi', 'expenses']].head()

print("Vectors:", (age_vector[:5], bmi_vector[:5]))
print("Factors:", (sex_factor[:5], smoker_factor[:5]))
print("List:", example_list)
print("Matrix:\n", matrix_example)
print("Data Frame:\n", subset_df)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(insurance_data['bmi'], insurance_data['expenses'], c='blue', marker='+')
plt.title('Scatter Plot of BMI vs Expenses')
plt.xlabel('BMI')
plt.ylabel('Expenses')
plt.text(30, 50000, 'High Expenses', fontsize=9, color='red')
plt.grid(True)
plt.show()
```

```python
smokers_data = insurance_data[insurance_data['smoker'] == 'yes']

plt.figure(figsize=(10, 6))
plt.scatter(smokers_data['bmi'], smokers_data['expenses'], c='red', marker='+')
plt.title('Scatter Plot of BMI vs Expenses (Smokers Only)')
plt.xlabel('BMI')
plt.ylabel('Expenses')
plt.text(30, 50000, 'High Expenses for Smokers', fontsize=9, color='blue')
plt.grid(True)
plt.show()
```

```python
smokers_data = insurance_data[insurance_data['smoker'] == 'no']

plt.figure(figsize=(10, 6))
plt.scatter(insurance_data['bmi'], insurance_data['expenses'], c='blue', marker='*', label='Non-Smokers')
plt.title('Scatter Plot of BMI vs Expenses (No Smokers Only)')
plt.xlabel('BMI')
plt.ylabel('Expenses')
plt.text(30, 25000, 'High Expenses for Smokers', fontsize=9, color='red')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(insurance_data['age'], insurance_data['expenses'], c='blue', marker='+', label='Original Data')
plt.title("Scatter Plot of Age vs Medical Expenses", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Medical Expenses", fontsize=12)
plt.text(30, 50000, "Higher expenses often occur for older individuals.", fontsize=10, color='green')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

```python
filtered_data=insurance_data.loc[(insurance_data['age']>=30) & (insurance_data['age']<=40)]
print(filtered_data)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['age'], filtered_data['expenses'], c='red', marker='o', label='Filtered Data')
plt.title("Scatter Plot of Age vs Medical Expenses (Filtered)", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Medical Expenses", fontsize=12)
plt.text(36, 52000, "Filtered data based upon 30>=age<=40", fontsize=10, color='green')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```
